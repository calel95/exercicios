```python
'''Ex 25 - Calcule as raízes da equação de 2º grau.
- A variável a tem que ser diferente de zero. Caso seja igual, imprima a mensagem "não é equação de segundo grau"
- Se Delte < 0, não existe real. Imprima a mensagem "Não existe raiz."
- Se Delta = 0, existe uma raiz real. Imprima a raiz e a mensagem "Raiz única."
- Se Delta > 0, imprima as duas raízes reais.'''
import math

A = int(input("A:"))
B = int(input("B:"))
C = int(input("C:"))
if A != 0:
    
    delta = B**2 - (4*A*C)
    print(delta)
    if delta <0:
        print("Não existe raízes reais.")
    elif delta ==0:
        x = (-B + math.sqrt(delta))/(A*2)
        print("Raiz única: ", x)
    elif delta > 0:
        x1 = (-B + math.sqrt(delta))/(A*2)
        x2 = (-B - math.sqrt(delta))/(A*2)
        print("Raiz X1: ", x1)
        print("Raiz X2: ", x2)
else:
    print("Não é uma equação do 2 grau.")
```

```python
'''Ex 28 Faça um programa que leia três números inteiros positivos e efetue o cálculo de uma das seguintes médias de acordo com
o valor numérico digitado pelo usuário:
a - Geométrica
b - Ponderada
c - Harmônica
d - Aritmétoca'''
print("ESCOLHA A LETRA DA MÉDIA QUE DESEJA:")
print("a - Geométrica\nb - Ponderada\nc - Harmônica\nd - Aritmétoca")
op = input()
op = op.lower()

n1 = float(input("N1: "))
n2 = float(input("N2: "))
n3 = float(input("N3: "))

if op == "a":
    media = (n1*n2*n3) **(1/3)
    print("Média Geométrica:", media)
elif op =="b":
    media = ((n1*1)+(2*n2)+(3*n3))/6
    print("Média Ponderada: ", media)
elif op == "c":
    media = 1/((1/n1)+(1/n2)+(1/n3))
    print("Média Harmônica:", media)
elif op == "d":
    media = (n1+n2+n3)/3
    print("Média Aritmética:", media)
else:
    print("OPÇÃO INVÁLIDA.")
```

```python
# Ex 41 - Faça um algoritmo que calcule o IMC de uma pessoa e mostre sua classificação de acordo com a tabela.
altura = float(input("Altura: "))
peso = float(input("Peso: "))

imc = peso / (altura*altura)
print("CLASSIFICAÇÂO DO SEU IMC:")
if imc <=18.5:
    print("Abaixo do peso: ", round(imc,2))
elif imc >18.5 and imc<25:
    print("Saudável: ", round(imc,2))
elif imc >=25 and imc<30:
    print("Peso em excesso: ", round(imc,2))
elif imc >=30 and imc <35:
    print("Obesidade Grau 1: ", round(imc,2))
elif imc >=35 and imc <40:
    print("Obesidade Grau 2 (severa): ", round(imc,2))
elif imc>=40:
    print("Obesidade Grau 3 (mórbida): ", round(imc,2))
```
